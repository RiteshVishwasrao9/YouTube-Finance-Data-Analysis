```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load cleaned data
df = pd.read_csv("cleaned_youtube_finance.csv")

# Top 10 most viewed videos
top_videos = df.sort_values(by="Views", ascending=False).head(10)
```

```python
# Bar Chart - Top 10 Videos by Views
plt.figure(figsize=(12, 6))
sns.barplot(y=top_videos["Title"], x=top_videos["Views"], palette="Blues_r")
plt.xlabel("Views (Millions)")
plt.ylabel("Video Title")
plt.title("Top 10 Trending Finance Videos")
plt.show()
```

```python
# Count of videos by popular channels
plt.figure(figsize=(12, 6))
sns.countplot(y=df["Channel"], order=df["Channel"].value_counts().index[:10], palette="coolwarm")
plt.title("Top 10 Finance Channels")
plt.xlabel("Number of Trending Videos")
plt.ylabel("Channel Name")
plt.show()
```
